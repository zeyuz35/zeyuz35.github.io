```python
# defining own functions

# without parameters
def function1():
  # print line
  print("function 1")
function1()

# indentation matters in python!
# functions can call other functions!
```

```python
# parameters and arguments
# as expected, functios do not pollute the global environment
def function2(x):
  sqr = x**2
  y = 2
  # this should be a return, but currently leave it as print for now
  print(sqr)
function2(x = 2)
x
y

# default parameters
def function3(x = 2):
  sqr = x**2
  print(sqr)
function3()

# named parameters, and mixing up the order
# this is totally fine as well, same as R
def function4(x = 2, y = 2):
  sqr = x*y
  print(sqr, x, y)
function4(y = 3, x = 2)

```

Scope, new content!
Actually not new, just the concept that functions don't leak into global environments

```python
# slightly new content
# if in the unlikely scenario you wish to actually change something in global
# prefix it with global keyword
def function5():
  global short 
  short = 1
  print(short)
function5()
short
# Note that R also does have this, but is NOT recommended
# x <<- 1
# this makes debugging an absolute nightmare
```
